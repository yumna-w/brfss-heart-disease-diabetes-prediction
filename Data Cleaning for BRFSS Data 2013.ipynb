(ii) Calculate missing variables

Our target variable, _MICHD, is present in the dataset for the year 2015 but needs to be calculated based on existing variables for the year 2013.

```python
# Create a new column '_MICHD' and calculate its values based on conditions
brfss_2013_df_selected['_MICHD'] = (brfss_2013_df_selected['CVDINFR4'] == 1) & (brfss_2013_df_selected['CVDCRHD4'] == 1)
# Convert True/False values to 1/0
brfss_2013_df_selected['_MICHD'] = brfss_2013_df_selected['_MICHD'].astype(int)
```

One feature, _RFDRHV5, is present in the dataset for the year 2015 but needs to be calculated based on existing variables for the year 2013.

```python
# Create a new column 'DROCDY3_' and initialize it with NaN values
brfss_2013_df_selected['DROCDY3_'] = float('nan')

# Apply the conditions and assign values to 'DROCDY3_'
brfss_2013_df_selected.loc[~brfss_2013_df_selected['ALCDAY5'].isin([888, 777, 999, pd.NaT]), 'DROCDY3_'] = brfss_2013_df_selected.loc[
    (brfss_2013_df_selected['ALCDAY5'].between(101, 107)), 'ALCDAY5'].sub(100).div(7)

brfss_2013_df_selected.loc[~brfss_2013_df_selected['ALCDAY5'].isin([888, 777, 999, pd.NaT]), 'DROCDY3_'] = brfss_2013_df_selected.loc[
    (brfss_2013_df_selected['ALCDAY5'].between(201, 230)), 'ALCDAY5'].sub(200).div(30)

brfss_2013_df_selected.loc[brfss_2013_df_selected['ALCDAY5'] == 888, 'DROCDY3_'] = 0
brfss_2013_df_selected.loc[~brfss_2013_df_selected['ALCDAY5'].isin([888, 777, 999, pd.NaT]), 'DROCDY3_'] = 9
```

```python
# Create a new column '_DRNKWEK' and initialize it with NaN values
brfss_2013_df_selected['_DRNKWEK'] = float('nan')

# Apply the conditions and assign values to '_DRNKWEK'
brfss_2013_df_selected.loc[brfss_2013_df_selected['DROCDY3_'] == 0, '_DRNKWEK'] = 0
brfss_2013_df_selected.loc[brfss_2013_df_selected['DROCDY3_'] == 9, '_DRNKWEK'] = 999
brfss_2013_df_selected.loc[brfss_2013_df_selected['AVEDRNK2'].isin([pd.NaT, 77, 99]), '_DRNKWEK'] = 999
brfss_2013_df_selected.loc[~brfss_2013_df_selected['AVEDRNK2'].isin([pd.NaT, 77, 99]), '_DRNKWEK'] = brfss_2013_df_selected['AVEDRNK2'] * brfss_2013_df_selected['DROCDY3_'] * 7
```

```python
# Create a new column '_RFDRHV5' and initialize it with NaN values
brfss_2013_df_selected['_RFDRHV5'] = float('nan')

# Apply the conditions and assign values to '_RFDRHV5'
brfss_2013_df_selected.loc[(brfss_2013_df_selected['SEX'] == 1) & (~brfss_2013_df_selected['_DRNKWEK'].isin([999, pd.NaT])), '_RFDRHV5'] = \
    brfss_2013_df_selected.loc[(brfss_2013_df_selected['SEX'] == 1) & (~brfss_2013_df_selected['_DRNKWEK'].isin([999, pd.NaT])), '_DRNKWEK'].apply(lambda x: 2 if x > 14 else 1)

brfss_2013_df_selected.loc[(brfss_2013_df_selected['SEX'] == 2) & (~brfss_2013_df_selected['_DRNKWEK'].isin([999, pd.NaT])), '_RFDRHV5'] = \
    brfss_2013_df_selected.loc[(brfss_2013_df_selected['SEX'] == 2) & (~brfss_2013_df_selected['_DRNKWEK'].isin([999, pd.NaT])), '_DRNKWEK'].apply(lambda x: 2 if x > 7 else 1)

brfss_2013_df_selected.loc[brfss_2013_df_selected['ALCDAY5'] == 888, '_RFDRHV5'] = 1

brfss_2013_df_selected.loc[brfss_2013_df_selected['_DRNKWEK'].isin([999, pd.NaT]), '_RFDRHV5'] = 9

brfss_2013_df_selected._RFDRHV5.unique()
```

```python
# Apply rounding to 'DROCDY3_' column
brfss_2013_df_selected['DROCDY3_'] = round(brfss_2013_df_selected['DROCDY3_'] * 100, 1)

# Apply rounding to '_DRNKWEK' column
brfss_2013_df_selected['_DRNKWEK'] = round(brfss_2013_df_selected['_DRNKWEK'] * 100, 1)
```
